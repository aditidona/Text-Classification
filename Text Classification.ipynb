## Creating a dictionary of all useful words from each document

```python
vocabulary={} # Dictionary of useful words

# Removed all punctuatons,stopwords, words of length < 2
for i in x_train:
    for j in i.split():
        j=j.strip(string.punctuation) 
        if j.lower() not in stopwords and len(j)>2 and not j.isdigit(): 
            if j not in vocabulary:
                vocabulary[j]=1
            else:
                vocabulary[j]+=1
```

```python
# Sorting the dictionary according to frequency of words in decreasing order
vocabulary= dict(sorted(vocabulary.items(), key = lambda kv:(kv[1], kv[0]), reverse=True))
```

## Finding the cut off frequency

```python
# creating an array to find the count of each frequency i.e num of words which have say 'f' frequency
num_words = [0 for i in range(max(vocabulary.values())+1)] 
freq = [i for i in range(max(vocabulary.values())+1)] 
for key in vocabulary:
    num_words[vocabulary[key]]+=1
plt.plot(freq,num_words)
plt.axis([1, 100, 0, 300]) 
plt.xlabel("No. of Words")
plt.ylabel("Frequency")
plt.grid()
plt.show()

# we can see that after freq=100, the num of words increase at a higher rate then their freq. 
# so we set cut off at 100
```

```python
# setting the cut_off frequency at 100
num_words=np.array(num_words)
cut_off=100
num_words_above_cut_off=num_words[cut_off:].sum()  # finding number of words greater than cut_off
print('Num of words which have occured in more than 100 times are: ',num_words_above_cut_off)
```

```python
# creating a list (FEATURES) selecting only those values which have frequency less than the cut off
features=np.array(list(vocabulary.keys())[0:num_words_above_cutt_off])
features
```

```python
# associating a feature number to each feature in the features list and creating a dictionary
vocab={}
for i in range(len(features)):
    vocab[features[i]]=i
```

## Creating Feature Matrix

```python
# FUNCTION TO CREATE A FEATURE_MATRIX CONTAINING THE COUNT OF EACH FEATURE IN THE DOCUMENTS DATASET
# ROWS->DOCUMENTS
# COLUMNS-> FEATURE NUMBERS
# WE READ EACH DOCUMENT AND INCREMENT ITS COUNT BY 1 CORRESPONDING THE FEATURE NUMBER 
def return_feature_matrix(dataset,features):
    feature_matrix=np.zeros((len(dataset),len(features)),int) #CREATING A LIST OF ZERO 
    for i in range(len(dataset)): 
        for j in dataset[i].split():
            if j in features:
                feature_matrix[i][vocab[j]]+=1
    return feature_matrix
```

```python
# creating feature matrix of x_train and x_test
x_train_matrix=return_feature_matrix(x_train,features)
x_test_matrix=return_feature_matrix(x_test,features)
```

# Implementation of Multinomial Naive Bayes from scratch

```python
class MultinomialNaiveBayes:
    
    def __init__(self):
        self.count = {} #dictionary which keeps the count of words of each class
    
    # FIT FUNCTION
    
    def fit(self,x_train,y_train):
        self.count['total_data']=len(y_train) #total number of data points
        classes=set(y_train)
        for current_class in classes:
            self.count[current_class]={}  #for each class make a dictionary of count of each word 
            current_class_rows=y_train==current_class
            x_train_current=x_train[current_class_rows] #select the current rows
            y_train_current=y_train[current_class_rows]
            self.count[current_class]['total_points']=len(x_train_current) # total number of data points belonging to curr class
            total=0
            for j in range(len(x_train[0])):
                total+=x_train_current[:,j].sum()
                self.count[current_class][j]=x_train_current[:,j].sum() # count of all words of that feature
            self.count[current_class]['total_freq']=total # this stores the total number of words in the current_class
    
    # FUNCTION TO FIND PROBABILITY THAT THE CURRENT DOCUMENT BELONGS TO CURRENT CLASS
    # using log to eliminate the possibility of value reducing to zero
    def probability_using_log(self,x,current_class):
        prob_current_class=np.log(self.count[current_class]['total_points'])-np.log(self.count['total_data']) # P(y==ai)
        output=0
        for j in range(len(x)):
            if x[j]!=0:
                count_current_class_words_with_word_j=self.count[current_class][j] + 1 # P(X=xj) given class==cur_class
                count_current_class_words=self.count[current_class]['total_freq'] + len(x) 
                current_xj_prob=np.log(count_current_class_words_with_word_j)-np.log(count_current_class_words)
                output+=(current_xj_prob + prob_current_class)*x[j]
        return output
    
    
    # FUNCTION TO CLASSIFY A SINGLE TEST DATA POINT
    def predict_one_point(self,x):
        classes=self.count.keys()
        best_p=-1000
        best_class=-1
        first=True
        for current_class in classes: #find probabbilites of this feature row being current class
            if current_class=="total_data":
                continue
            p_current=self.probability_using_log(x,current_class) #calc porbability that document belongs to curr_class
            if first or p_current>best_p:
                best_p=p_current
                best_class=current_class
            first=False
        return best_class #predict the max out of all

    # FUNCTION TO PREDICT  
    def predict(self,x_test):
        y_pred=[]
        for x in x_test:
            temp=self.predict_one_point(x) #sending one class to for prediction 
            y_pred.append(temp)
        return y_pred
```

```python
clf2 = MultinomialNaiveBayes()
clf2.fit(x_train_matrix,y_train)
y_pred2=clf2.predict(x_test_matrix)
print('--------------CLASSIFICATION REPORT-----------------')
print(classification_report(y_test,y_pred2))
print('--------------CONFUSION MATRIX----------------------')
print(confusion_matrix(y_test,y_pred2))
```

## Comparison of the two implementations

```python
# Score function to check the mean accuracy
def score(Y_pred,Y_true):
    # returns the mean accuracy
    count = 0
    for i in range(len(Y_pred)):
        if Y_pred[i] == Y_true[i]:
            count+=1
    return count/len(Y_pred)
```
